# 4. Análise Univariada

```python
# Idade
_ = plt.hist(candidatos.NR_IDADE_DATA_POSSE)
plt.show()
```

```python
# Gênero
candidatos.DS_GENERO.value_counts(1)
```

Em relação ao gênero, cerca de 68% dos candidatos declaram-se gênero masculino, frente a cerca de 32% dos candidatos de gênero feminino.

```python
# Grau de Instrução
candidatos.DS_GRAU_INSTRUCAO.value_counts(1)
```

```python
# Grau de Instrução
escolaridade = candidatos.DS_GRAU_INSTRUCAO.value_counts(1)
```

```python
escolaridade.plot.bar()
_ = plt.title('Grau de Instrução dos Candidatos - RJ')
plt.grid(linestyle = '--', linewidth = 0.5)
plt.show()
```

Vemos que quase cerca de metade dos candidatos apresenta grau de escolaridade com ensino superior completo, seguido do grau de escolaridade de ensino médio completo.

```python
# Raça

candidatos.DS_COR_RACA.value_counts(1).plot.bar()
_ = plt.title('Cor/Raça dos Candidatos - RJ')
plt.show()
```

Quanto à raça, cerca de 50% dos candidatos se declaram brancos, seguidos por pardos e pretos.

# 4. Análise Bivariada

```python
gen_escolaridade = candidatos.groupby(['DS_GENERO', 'DS_GRAU_INSTRUCAO']).size().unstack(1)
gen_escolaridade.head()
```

```python
cor_escolaridade = candidatos.groupby(['DS_COR_RACA', 'DS_GRAU_INSTRUCAO']).size().unstack(1)


cor_escolaridade['TOTAL'] = cor_escolaridade.sum(axis=1)

cor_escolaridade['ENS FUND C %'] = round(cor_escolaridade['ENSINO FUNDAMENTAL COMPLETO']/cor_escolaridade['TOTAL']*100)
cor_escolaridade['ENS FUND INC %'] = round(cor_escolaridade['ENSINO FUNDAMENTAL INCOMPLETO']/cor_escolaridade['TOTAL']*100)
cor_escolaridade['ENS MEDIO C %'] = round(cor_escolaridade['ENSINO MÉDIO COMPLETO']/cor_escolaridade['TOTAL']*100)
cor_escolaridade['ENS MEDIO INC %'] = round(cor_escolaridade['ENSINO MÉDIO INCOMPLETO']/cor_escolaridade['TOTAL']*100)
cor_escolaridade['LÊ E ESCREVE %'] = round(cor_escolaridade['LÊ E ESCREVE']/cor_escolaridade['TOTAL']*100)
cor_escolaridade['SUP C %'] = round(cor_escolaridade['SUPERIOR COMPLETO']/cor_escolaridade['TOTAL']*100)
cor_escolaridade['SUP INC %'] = round(cor_escolaridade['SUPERIOR INCOMPLETO']/cor_escolaridade['TOTAL']*100)

cor_escolaridade
```

Observa-se que há diferença entre  o grau de instrução dos candidatos em relação a cor/raça. Os candidatos que se declaram branco em sua maioria possuem Ensino Superior Completo, seguidos de Ensino Médio Completo. Enquanto os candidatos pardos e pretos, a maioria possui ensino médio completo, seguido de ensino superior completo. Dos pouco candidatos que se declaram indígenas, a maioria possui ensino superior completo. E os dois candidatos de cor amarela, ambos possuem superior completo.
